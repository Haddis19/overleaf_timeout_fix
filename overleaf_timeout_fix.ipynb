Ste 2:create workspace

```python
import os

BASE_DIR = "/content/latexprojects"
os.makedirs(BASE_DIR, exist_ok=True)

print("📁 Workspace ready:", BASE_DIR)
```

Step 3: Upload Project Files Upload your .tex file (required). If your document uses images, bibliography files, or other supporting files, upload them here as well.

```python
from google.colab import files
import os, shutil

uploaded = files.upload()

existing_files = set(os.listdir(BASE_DIR))

for f in uploaded.keys():
    target_path = os.path.join(BASE_DIR, f)

    if f in existing_files:
        raise FileExistsError(f"❌ File '{f}' already exists!")

    shutil.move(f, target_path)

print("✅ Upload successful (stored locally)")
```

Step 4: Select a LaTeX File


```python

import glob
import os
import ipywidgets as widgets
from IPython.display import display

tex_files = glob.glob(os.path.join(BASE_DIR, "*.tex"))

if len(tex_files) == 0:
    raise FileNotFoundError(
        "❌ No .tex files found.\nUpload a .tex file first."
    )

options = {os.path.basename(f): f for f in sorted(tex_files)}

dropdown = widgets.Dropdown(
    options=options,
    description="Select .tex:",
    layout=widgets.Layout(width='60%')
)

print(f"📄 Found {len(tex_files)} .tex file(s)")
display(dropdown)
```

Step 5: Compile the Document

```python
import subprocess
import os

BASE_DIR = "/content/latexprojects"

def compile_latex(tex_file):
    tex_name = os.path.basename(tex_file)

    cmd = [
        "xelatex",
        "-interaction=nonstopmode",
        tex_name
    ]

    result1 = subprocess.run(
        cmd,
        cwd=BASE_DIR,
        capture_output=True,
        text=True
    )

    result2 = subprocess.run(
        cmd,
        cwd=BASE_DIR,
        capture_output=True,
        text=True
    )

    if result2.returncode != 0:
        print("❌ Compilation failed\n")
        print(result2.stdout)
        print(result2.stderr)
        return False

    print("✅ Compilation finished successfully")
    return True


tex_file = dropdown.value
tex_name = os.path.splitext(os.path.basename(tex_file))[0]

success = compile_latex(tex_file)

if success:
    pdf_file = os.path.join(BASE_DIR, tex_name + ".pdf")
    print("📄 PDF saved at:")
    print(pdf_file)
```

Step 6: Download the PDF

```python
from google.colab import files
import os

BASE_DIR = "/content/latexprojects"

pdf_file = os.path.join(BASE_DIR, tex_name + ".pdf")

if not os.path.exists(pdf_file):
    raise FileNotFoundError(
        "❌ PDF not found. Compilation may have failed."
    )

print("📄 Downloading:", os.path.basename(pdf_file))

files.download(pdf_file)
```
